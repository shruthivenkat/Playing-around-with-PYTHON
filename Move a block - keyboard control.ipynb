```python
import pygame as pg
```

```python
pg.init()
```

```python
screen = pg.display.set_mode((400,300))
```

```python
done = False
is_blue = True
x = 40
y = 30
clock = pg.time.Clock()
while not done:
    
    for event in pg.event.get():
        if event.type == pg.QUIT:
            done = True
        if event.type == pg.KEYDOWN and event.key == pg.K_SPACE:
            is_blue = not is_blue
    
    pressed = pg.key.get_pressed()
    if pressed[pg.K_UP]:
        y -= 3
    if pressed[pg.K_DOWN]:
        y += 3
    if pressed[pg.K_LEFT]:
        x -= 3
    if pressed[pg.K_RIGHT]:
        x += 3
        
    if is_blue:
        color = (0, 120, 255)
    else:
        color = (255, 100, 0)
    screen.fill((0,0,0)) # ----->  this will erase the previously created rectangle
    pg.draw.rect(screen, color, pg.Rect(x,y,12,12))
#     screen.fill((0,0,0))
    clock.tick(60) #------> slows down the fastly moving image. the imgae moves at 1/60th speed. or sometimes, 1/1000. so, we use a clock to capture the imgae
    
    pg.display.flip()
```
